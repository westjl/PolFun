```python
%matplotlib ipympl
import numpy as np

import matplotlib
import os
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from mpl_toolkits.mplot3d import Axes3D
import mpl_toolkits.mplot3d.art3d as art3d
from matplotlib import gridspec
import pdb

import networkx as nx
import matplotlib.animation as anim
```

```python
ne=0. #cm-3
e=4.8E-10 #esu.
me=9.1094E-28 #g
B=5E-6 #G
c=3E10 #cm/s
f=2E4 #Hz
wave = c/f
print(str(wave)+" cm")

##Use Faraday rotation
omega_e = np.sqrt(4*np.pi*ne*e*e/me) #Hz => 25kHz
omega_e_sq = omega_e*omega_e
Omega_e = -e*B/me/c #


ang_freq = f/2/np.pi

k = 2*np.pi / wave

omega_sq_RCP = c*c*k*k + omega_e_sq + (omega_e_sq * np.abs(Omega_e) / ang_freq)
omega_sq_LCP = c*c*k*k + omega_e_sq - (omega_e_sq * np.abs(Omega_e) / ang_freq)

denom_RCP = 1 - (np.abs(Omega_e) / ang_freq)
denom_LCP = 1 + (np.abs(Omega_e) / ang_freq)

omega_sq_RCP = c*c*k*k + (omega_e_sq / denom_RCP)
omega_sq_LCP = c*c*k*k + (omega_e_sq / denom_LCP)

t2=omega_e_sq/2/(ang_freq*ang_freq)
t3=omega_e_sq* np.abs(Omega_e)/2/(ang_freq*ang_freq*ang_freq)

V_RCP = c* (1+t2+t3)
V_LCP = c* (1+t2-t3)

omega_RCP = np.sqrt(omega_sq_RCP)
omega_LCP = np.sqrt(np.abs(omega_sq_LCP))

k_RCP = omega_RCP/V_RCP
k_LCP = omega_LCP/V_LCP

print ("V_RCP = {:.2e}".format(V_RCP))
print ("V_LCP = {:.2e}".format(V_LCP))
print ("omega_RCP = {:.2e}".format(omega_RCP))
print ("omega_LCP = {:.2e}".format(omega_LCP))
print ("k_RCP = {:.2e}".format(k_RCP) )
print ( "k_LCP = {:.2e}".format(k_LCP))
print ("1/k_RCP = {:.4e}".format(1/k_RCP) )
print ( "1/k_LCP = {:.4e}".format(1/k_LCP))

#s = distance from the source, source at s=0
#With the second definition, ùúô at a given time ùë° decreases with increasing distance ùë†, 
#while ùúô at a given distance ùë† increases with increasing time ùë°; 
#this means that ùúô increases in the same sense as the rotation of the electric field vector, 
#such that ùúô can be taken equal to the angle by which the electric field vector has rotated 
#(since the time when ùúô = 0)

t=0
s=0

phase_RCP= omega_RCP * t - k_RCP * s
phase_LCP= omega_LCP * t - k_LCP * s


phase_diff = 0
angle_th = np.pi/4

E1 = np.cos(angle_th)
E2 = np.sin(angle_th)

#f=10Hz
tmin=0
tmax=1/f
zmin=-10/k_RCP
zmax=10/k_RCP

print (zmax)
```
